```python
import daft
import matplotlib.pyplot as plt
```

# Plate Diagram (style and topic independent)

```python
plate_diag = daft.PGM()
```

```python
plate_diag.add_plate([0, 0, 3, 2], label = r'M')

# Nodes
plate_diag.add_node('d', r'd', 2, 1.5, observed = True)
plate_diag.add_node('s', r's', 2, .5, observed = True)

plate_diag.add_node('theta', r'$\theta$', 1, 1.5)
plate_diag.add_node('kappa', r'$\kappa$', 1, .5)

plate_diag.add_node('beta', r'$\beta$', 1.5, 2.6)
plate_diag.add_node('gamma', r'$\gamma$', 1.5, -0.6)

plate_diag.add_node('eta_beta', r'$\eta_\beta$', .5, 2.6)
plate_diag.add_node('eta_gamma', r'$\eta_\gamma$', .5, -0.6)

plate_diag.add_node('alpha_theta', r'$\alpha_\theta$', -0.6, 1.5)
plate_diag.add_node('alpha_kappa', r'$\alpha_\kappa$', -0.6, .5)

# Edges
plate_diag.add_edge('theta', 'd')
plate_diag.add_edge('kappa', 's')

plate_diag.add_edge('beta', 'd')
plate_diag.add_edge('gamma', 's')

plate_diag.add_edge('eta_beta', 'beta')
plate_diag.add_edge('eta_gamma', 'gamma')

plate_diag.add_edge('alpha_theta', 'theta')
plate_diag.add_edge('alpha_kappa', 'kappa')
```

```python
plate_diag.render()
```
