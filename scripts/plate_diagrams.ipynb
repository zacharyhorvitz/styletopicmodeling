```python
import daft
import matplotlib.pyplot as plt
```

# Plate Diagram (style and topic independent)

```python
style_pparams = {'edgecolor': 'blue', 'color': 'blue'}
style_fpparams = dict({'facecolor': 'blue'}, **style_pparams)
```

```python
plate_diag = daft.PGM()

plate_diag.add_plate([0, 0, 3, 2], label = r'M')
plate_diag.add_plate([.5, 2.2, 1, .8], label = r'K')
plate_diag.add_plate([.5, -1, 1, .8], label = r'J')

plate_diag.add_plate([.5, .1, 2, .75], label = r'S')

# Nodes
plate_diag.add_node('d', r'd', 2, 1.5, observed = True)
plate_diag.add_node('s', r's', 2, .5, observed = True, plot_params = style_pparams)

plate_diag.add_node('theta', r'$\theta$', 1, 1.5)
plate_diag.add_node('kappa', r'$\kappa$', 1, .5, plot_params = style_pparams)

plate_diag.add_node('beta_theta', r'$\beta^\theta$', 1, 2.6)
plate_diag.add_node('gamma', r'$\gamma$', 1, -0.6, plot_params = style_pparams)

plate_diag.add_node('alpha_theta', r'$\alpha^\theta$', -0.6, 1.5, fixed = True)
plate_diag.add_node('alpha_kappa', r'$\alpha^\kappa$', -0.6, .5, fixed = True, plot_params = style_fpparams)

# Edges
plate_diag.add_edge('theta', 'd')
plate_diag.add_edge('kappa', 's', plot_params = style_fpparams)

plate_diag.add_edge('beta_theta', 'd')
plate_diag.add_edge('gamma', 's', plot_params = style_fpparams)

plate_diag.add_edge('alpha_theta', 'theta')
plate_diag.add_edge('alpha_kappa', 'kappa', plot_params = style_fpparams)
```

```python
plate_diag.savefig('indep_plate_diagram.png', dpi = 500)
plate_diag.render()
```

# Plate Diagram (style influence topic)

```python
style_pparams = {'edgecolor': 'blue', 'color': 'blue'}
style_fpparams = dict({'facecolor': 'blue'}, **style_pparams)

link_pparams = {'edgecolor': 'red', 'color': 'red'}
link_fpparams = dict({'facecolor': 'red'}, **link_pparams)
```

```python
plate_diag = daft.PGM()

plate_diag.add_plate([0, 0, 3, 2], label = r'M')
plate_diag.add_plate([.5, 2.2, 1, .8], label = r'K')
plate_diag.add_plate([.5, -1, 1, .8], label = r'J')
plate_diag.add_plate([1.6, 2.2, 1, .8], label = r'J')

plate_diag.add_plate([.5, .1, 2, .75], label = r'S')

# Nodes
plate_diag.add_node('d', r'd', 2, 1.5, observed = True)
plate_diag.add_node('s', r's', 2, .5, observed = True, plot_params = style_pparams)

plate_diag.add_node('theta', r'$\theta$', 1, 1.5)
plate_diag.add_node('kappa', r'$\kappa$', 1, .5, plot_params = style_pparams)

plate_diag.add_node('beta_theta', r'$\beta^\theta$', 1, 2.6)
plate_diag.add_node('beta_kappa', r'$\beta^\kappa$', 2.1, 2.6, plot_params = link_pparams)
plate_diag.add_node('gamma', r'$\gamma$', 1, -0.6, plot_params = style_pparams)

plate_diag.add_node('alpha_theta', r'$\alpha^\theta$', -0.6, 1.5, fixed = True)
plate_diag.add_node('alpha_kappa', r'$\alpha^\kappa$', -0.6, .5, fixed = True, plot_params = style_fpparams)

# Edges
plate_diag.add_edge('theta', 'd')
plate_diag.add_edge('kappa', 's', plot_params = style_fpparams)
plate_diag.add_edge('kappa', 'd', plot_params = link_fpparams)

plate_diag.add_edge('beta_theta', 'd')
plate_diag.add_edge('beta_kappa', 'd', plot_params = link_fpparams)
plate_diag.add_edge('gamma', 's', plot_params = style_fpparams)

plate_diag.add_edge('alpha_theta', 'theta')
plate_diag.add_edge('alpha_kappa', 'kappa', plot_params = style_fpparams)
```

```python
plate_diag.savefig('link_plate_diagram.png', dpi = 500)
plate_diag.render()
```
